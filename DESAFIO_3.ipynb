```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from yellowbrick.cluster import KElbowVisualizer
import numpy as np
import matplotlib.pyplot as plt

#Carregando dataset por meio do pandas
df = pd.read_csv('/content/Failing_Equipment_Exercise.csv')
df = df.transpose()
colunas = df.columns
score = []

#Loop onde em cada iteração um sensor é retirado do modelo e verificado seu desempenho em sua ausência
for coluna in colunas:
  _df = df.drop(columns=[coluna])
  X, y = _df.to_numpy(), np.array(_df.index)

  x_linha_adicionar = np.array([X[0]])
  _X = np.append(X, x_linha_adicionar, 0)
  y_linha_adicionar = np.array(y[0])
  _y = np.append(y, y_linha_adicionar)

#Replicação das instâncias do dataset para o modelo poder processar os dados

  for index in range(1000):
    x_linha_adicionar = np.array([X[1]])
    _X = np.append(_X, x_linha_adicionar, 0)
    y_linha_adicionar = np.array(y[1])
    _y = np.append(_y, y_linha_adicionar)
    x_linha_adicionar = np.array([X[2]])
    _X = np.append(_X, x_linha_adicionar, 0)
    y_linha_adicionar = np.array(y[2])
    _y = np.append(_y, y_linha_adicionar)
    x_linha_adicionar = np.array([X[3]])
    _X = np.append(_X, x_linha_adicionar, 0)
    y_linha_adicionar = np.array(y[3])
    _y = np.append(_y, y_linha_adicionar)
    x_linha_adicionar = np.array([X[0]])
    _X = np.append(_X, x_linha_adicionar, 0)
    y_linha_adicionar = np.array(y[0])
    _y = np.append(_y, y_linha_adicionar)

  #Implementação do modelo de clusterização e do "Elbow Method"
  model = KMeans()
  visualizer = KElbowVisualizer(
      model, k=(2,12), metric='calinski_harabasz', timings=False
  )

  visualizer.fit(_X)       
  visualizer.show()        
  score.append(visualizer.elbow_score_)
  print(_df.columns)
  print(_X.shape)
  print(_y.shape)
```

```python
#Relatório Desafio 3

# O "Elbow Method" é uma técnica utilizada para auxiliar cientistas de dados a determinar 
# o número ideal de clusters, ajustando o modelo com um intervalo de valores K, e com o
# o gráfico possuindo o traçado que se assemelha a um braço, será no cotovelo ou ponto
# de inflexão da curva um indicativo de que o modelo tem melhor representação naquele
# ponto.
# Este trabalho é implementado por meio do método de clusterização "KMeans" da biblioteca sklearn 
# que é utilizado para gerar o modelo e também com o método "KElbowVisualizer" fornecido pela biblioteca 
# Yellowbrick que auxilia na plotagem de dados e cálculo das métricas do "Elbow Method", por padrão a
# biblioteca utiliza como métrica a "distortion", que representa a soma das distâncias quadradas de cada 
# ponto ao centro atribuído, entretanto no contexto do projeto foi adotada a métrica "calinski_harabasz",
# que determina o score calculando a razão de dispersão entre e dentro dos clusters.
# A partir do dataset dado, temos como parâmetros 17 sensores e 4 "máquinas", e devemos determinar qual sensor
# opera em defeito. Assumindo como hipotése que o sensor com defeito impacta negativamente no score da clusterização
# dos dados, podemos implementar uma iteração de clusterizações, onde em cada iteração excluímos a presença de um sensor
# a iteração em que obtivermos o melhor score, será onde o sensor que apresenta falha foi retirado, logo pelo gráfico de
# "Análise do Data Set por Elbow Method" temos que o sensor 4 é o que impacta o modelo com o melhor score quando retirado,
# sendo portanto o sensor 4 o defeituoso, outro aspecto interessante é que o modelo aponta para a existência de 3 clusters,
# logo aparentemente, entre as 4 máquinas temos duas máquinas iguais.

```

```python
plt.title("Análise do Data Set por Elbow Method")
plt.xlabel("Sensores retirados para o teste")
plt.ylabel("Calinski Harabasz Score")
plt.grid(True)
plt.plot(colunas, score, color='blue', label = 'Acurácia da rede Neural Convolucional')
plt.scatter(colunas[score .index(max(score))], max(score), color='red', label='Valor Máximo Score val.: %0.4f' % max(score))
plt.legend()
plt.xticks(rotation=45)
plt.savefig('Acurácia de rede Neural Convolucional com kernel 9x9 e 1024 filtros.png', dpi=1080, facecolor='w', edgecolor='w', orientation='landscape', papertype=None, format=None,transparent=False, bboxinches=None, padinches=0.1,frameon=None, bbox_inches='tight')
plt.show()
```
