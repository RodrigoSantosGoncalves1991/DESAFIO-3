```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from yellowbrick.cluster import KElbowVisualizer
import numpy as np
import matplotlib.pyplot as plt

#Carregando dataset por meio do pandas
df = pd.read_csv('/content/Failing_Equipment_Exercise.csv')
df = df.transpose()
colunas = df.columns
score = []

#Loop onde em cada iterao um sensor  retirado do modelo e verificado seu desempenho em sua ausncia
for coluna in colunas:
  _df = df.drop(columns=[coluna])
  X, y = _df.to_numpy(), np.array(_df.index)

  x_linha_adicionar = np.array([X[0]])
  _X = np.append(X, x_linha_adicionar, 0)
  y_linha_adicionar = np.array(y[0])
  _y = np.append(y, y_linha_adicionar)

#Replicao das instncias do dataset para o modelo poder processar os dados

  for index in range(1000):
    x_linha_adicionar = np.array([X[1]])
    _X = np.append(_X, x_linha_adicionar, 0)
    y_linha_adicionar = np.array(y[1])
    _y = np.append(_y, y_linha_adicionar)
    x_linha_adicionar = np.array([X[2]])
    _X = np.append(_X, x_linha_adicionar, 0)
    y_linha_adicionar = np.array(y[2])
    _y = np.append(_y, y_linha_adicionar)
    x_linha_adicionar = np.array([X[3]])
    _X = np.append(_X, x_linha_adicionar, 0)
    y_linha_adicionar = np.array(y[3])
    _y = np.append(_y, y_linha_adicionar)
    x_linha_adicionar = np.array([X[0]])
    _X = np.append(_X, x_linha_adicionar, 0)
    y_linha_adicionar = np.array(y[0])
    _y = np.append(_y, y_linha_adicionar)

  #Implementao do modelo de clusterizao e do "Elbow Method"
  model = KMeans()
  visualizer = KElbowVisualizer(
      model, k=(2,12), metric='calinski_harabasz', timings=False
  )

  visualizer.fit(_X)       
  visualizer.show()        
  score.append(visualizer.elbow_score_)
  print(_df.columns)
  print(_X.shape)
  print(_y.shape)
```

```python
#Relatrio Desafio 3

# O "Elbow Method"  uma tcnica utilizada para auxiliar cientistas de dados a determinar 
# o nmero ideal de clusters, ajustando o modelo com um intervalo de valores K, e com o
# o grfico possuindo o traado que se assemelha a um brao, ser no cotovelo ou ponto
# de inflexo da curva um indicativo de que o modelo tem melhor representao naquele
# ponto.
# Este trabalho  implementado por meio do mtodo de clusterizao "KMeans" da biblioteca sklearn 
# que  utilizado para gerar o modelo e tambm com o mtodo "KElbowVisualizer" fornecido pela biblioteca 
# Yellowbrick que auxilia na plotagem de dados e clculo das mtricas do "Elbow Method", por padro a
# biblioteca utiliza como mtrica a "distortion", que representa a soma das distncias quadradas de cada 
# ponto ao centro atribudo, entretanto no contexto do projeto foi adotada a mtrica "calinski_harabasz",
# que determina o score calculando a razo de disperso entre e dentro dos clusters.
# A partir do dataset dado, temos como parmetros 17 sensores e 4 "mquinas", e devemos determinar qual sensor
# opera em defeito. Assumindo como hipotse que o sensor com defeito impacta negativamente no score da clusterizao
# dos dados, podemos implementar uma iterao de clusterizaes, onde em cada iterao exclumos a presena de um sensor
# a iterao em que obtivermos o melhor score, ser onde o sensor que apresenta falha foi retirado, logo pelo grfico de
# "Anlise do Data Set por Elbow Method" temos que o sensor 4  o que impacta o modelo com o melhor score quando retirado,
# sendo portanto o sensor 4 o defeituoso, outro aspecto interessante  que o modelo aponta para a existncia de 3 clusters,
# logo aparentemente, entre as 4 mquinas temos duas mquinas iguais.

```

```python
plt.title("Anlise do Data Set por Elbow Method")
plt.xlabel("Sensores retirados para o teste")
plt.ylabel("Calinski Harabasz Score")
plt.grid(True)
plt.plot(colunas, score, color='blue', label = 'Acurcia da rede Neural Convolucional')
plt.scatter(colunas[score .index(max(score))], max(score), color='red', label='Valor Mximo Score val.: %0.4f' % max(score))
plt.legend()
plt.xticks(rotation=45)
plt.savefig('Acurcia de rede Neural Convolucional com kernel 9x9 e 1024 filtros.png', dpi=1080, facecolor='w', edgecolor='w', orientation='landscape', papertype=None, format=None,transparent=False, bboxinches=None, padinches=0.1,frameon=None, bbox_inches='tight')
plt.show()
```
